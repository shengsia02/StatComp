#### $\textcolor{lightseagreen}{1.\ Normal\ Distribution}$

- $分佈:$ $$X \sim N(\mu, \sigma^{2}), \quad -\infty < \mu < \infty, \quad \sigma > 0$$

- $PDF:$ $$f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}, \quad -\infty < x < \infty$$

<hr>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize = (8, 6))
plt.suptitle('Standard Normal Distributions N(0, 1)', fontsize = 18, fontweight = 'bold')
plt.xlabel('$\mu$', fontsize = 15, color = 'black', fontweight = 'bold')
plt.ylabel('Density', fontsize = 15, color = 'black', fontweight = 'bold')

# 標準常態分配
xlim = [-5, 5]
mu = 0
s = 1
x = np.linspace(xlim[0], xlim[1], 1000)        # xlim[0] 為 xlim 的最小值，xlim[1] 為 xlim 的最大值

# 定义颜色列表
# colors = plt.colormaps['turbo'](np.linspace(0.3, 1, len(s)))


y = norm.pdf(x, mu, s)
ax.plot(x, y, linewidth = 3, color = 'blueviolet', alpha = 0.8)
ax.set_yticks(np.arange(0, 0.41, 0.1))
ax.tick_params(axis = 'both', colors = 'black') # 设置 x, y 轴刻度线的颜色
ax.legend(['$\mu$ = 0, $\sigma$ = 1'], fontsize = 13, loc = 'upper right')

plt.tight_layout()
plt.show()
```

#### $\textcolor{lightseagreen}{2.\ \chi^{2}\ Distribution}$

- $分佈:$ $$X \sim \chi^{2}(\nu), \quad \nu = 1, 2, ...$$

- $PDF:$ $$f(x) = \frac{1}{\Gamma(\frac{\nu}{2}) 2^{\frac{\nu}{2}}} x^{\frac{\nu}{2}-1} e^{-\frac{x}{2}}, \quad 0 < x < \infty$$

<hr>

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure(figsize = (7, 5))
fig = plt.gcf()  # get current figure（获取当前图形）
ax = plt.gca()  # get current axis（获取当前轴对象）

bg_color = plt.rcParams['axes.facecolor']
fig.patch.set_facecolor(bg_color)

xlim = [0, 50]
x = np.linspace(xlim[0], xlim[1], 1000)
dfs = np.arange(4, 32, 1)  # 自由度

# 定义颜色列表
colors = plt.colormaps['Blues'](np.linspace(0.5, 1, len(dfs)))

# fix xlim before animation
plt.axis([xlim[0], xlim[1], 0, 0.2])

for df, color in zip(dfs, colors):
    y = chi2.pdf(x, df)
    line, = plt.plot(x, y, lw = 1, color = color, label = r'$\nu$ = {}'.format(df))
    plt.xlabel('$d.f.$', fontsize = 12, color = 'black')
    plt.ylabel('Density', fontsize = 12, color = 'black')

    # 添加新的图例
    ax.legend([line], [line.get_label()], loc = 'upper right')  # 调整图例的位置

    [spine.set_visible(False) for spine in ax.spines.values()] # 隐藏子图的外框
    plt.grid(True, linestyle = '-', color = '#B6AF9F')
    plt.tick_params(axis = 'both', colors = 'black')  # 设置 x, y 轴刻度线的颜色
    plt.title('$\chi^2$ Distributions', fontsize = 13, fontweight='bold')
    plt.yticks([0, 0.1, 0.2])
    plt.tight_layout()
    # plt.pause(0.5)

    # 清除之前的图例
    if ax.get_legend() is not None:
        ax.get_legend().remove()

plt.show()
```

#### $\textcolor{lightseagreen}{3.\ T\ Distribution}$

- $分佈:$ $$X \sim T(\nu), \quad \nu = 1, 2, ...$$

- $PDF:$ $$f(x) = \frac{\Gamma(\frac{\nu+1}{2})}{\sqrt{\nu\pi}\Gamma(\frac{\nu}{2})} (1+\frac{x^{2}}{\nu})^{-\frac{\nu+1}{2}}, \quad -\infty < x < \infty$$

<hr>

```python
import numpy as np
from scipy.stats import t, norm
import matplotlib.pyplot as plt

plt.style.use('ggplot')
fig, ax = plt.subplots(figsize = (7, 5))

# 设置图形背景颜色
bg_color = plt.rcParams['axes.facecolor']
fig.patch.set_facecolor(bg_color)

xlim = [-6, 6]
x1 = np.linspace(xlim[0], xlim[1], 1000)
df1 = np.arange(0.1, 1.1, 0.1)
df2 = np.arange(3, 33, 3) # t 分佈自由度 >= 30 時，t 分佈近似於標準正態分佈
dfs = np.concatenate([df1, df2])

# 定义颜色列表
colors = plt.colormaps['Greens'](np.linspace(0.4, 1, len(dfs)))

# fix xlim before animation
ax.axis([xlim[0], xlim[1], 0, 0.4])

for df, color in zip(dfs, colors):
    y1 = t.pdf(x1, df)
    line, = ax.plot(x1, y1, linewidth = 1.5, color = color, label = r'$\nu$ = {}'.format(df))
    ax.legend([line], [line.get_label()], loc = 'upper right')
    [spine.set_visible(False) for spine in ax.spines.values()] # 隐藏子图的外框
    plt.grid(True, linestyle = '-', color = '#B6AF9F')
    ax.tick_params(axis = 'both', colors = 'black')
    ax.set_title('$t$ Distributions', fontsize = 15, fontweight = 'bold')
    ax.set_xlabel('$d.f.$', fontsize = 12, color = 'black')
    ax.set_ylabel('Density', fontsize = 12, color = 'black')
    ax.set_ylim([0, 0.41])
    ax.set_yticks(np.arange(0, 0.41, 0.1))
    plt.tight_layout()
    # plt.pause(0.5)

    # 清除之前的图例
    if ax.get_legend() is not None:
        ax.get_legend().remove()

x2 = np.linspace(xlim[0], xlim[1], 1000)
y2 = norm.pdf(x2, loc = 0, scale = 1)
line2, = ax.plot(x2, y2, color = 'red', linewidth = 2, label = 'Z Distribution')
ax.legend([line2], [line2.get_label()], loc = 'upper right')

plt.show()
```

#### $\textcolor{lightseagreen}{4.\ \beta\ Distribution}$

- $分佈:$ $$X \sim Beta(\alpha, \beta), \quad \alpha ,\; \beta > 0$$

- $PDF:$ $$f(x) = \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)+\Gamma(\beta)} x^{\alpha-1}(1-x)^{\beta-1}, \quad 0 < x < 1$$

<hr>

```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt

plt.style.use('ggplot')
fig, ax = plt.subplots(figsize = (7, 5))

# 设置图形背景颜色
bg_color = plt.rcParams['axes.facecolor']
fig.patch.set_facecolor(bg_color)

xlim = [0, 1]
x = np.linspace(xlim[0], xlim[1], 1000)
a = 9
b = np.arange(1, 31, 1)

# 定义颜色列表
colors = plt.colormaps['Oranges'](np.linspace(0.3, 1, len(b)))

# fix xlim before animation
ax.axis([xlim[0], xlim[1], 0, 8.8])

for b, color in zip(b, colors):
    y = beta.pdf(x, a = a, b = b)
    line, = ax.plot(x, y, linewidth = 1.5, color = color, label = r'$\alpha$ = {}, $\beta$ = {}'.format(a, b))
    ax.legend([line], [line.get_label()], loc = 'upper right')
    [spine.set_visible(False) for spine in ax.spines.values()] # 隐藏子图的外框
    plt.grid(True, linestyle = '-', color = '#B6AF9F')
    ax.tick_params(axis = 'both', colors = 'black')
    ax.set_title(r'$\beta$ Distributions', fontsize = 13, fontweight = 'bold')
    ax.set_xlabel('$x$', fontsize = 12, color = 'black')
    ax.set_ylabel('Density', fontsize = 12, color = 'black')
    ax.set_ylim([0, 8.8])
    plt.tight_layout()
    # plt.pause(0.5)

    # 清除之前的图例
    if ax.get_legend() is not None:
        ax.get_legend().remove()

plt.text(0.5, 6.5, r'$\alpha = 9$, $1 \leq \beta \leq 30$', fontsize = 12, color = 'k', ha = 'left')
plt.show()
```

#### $\textcolor{lightseagreen}{5.\ F\ Distribution}$

- $分佈:$ $$X \sim F(n_1, n_2), \quad n_1 ,\; n_2 > 0$$

- $PDF:$ $$f(x) = \frac{\Gamma(\frac{n_1+n_2}{2})}{\Gamma(\frac{n_1}{2})+\Gamma(\frac{n_2}{2})} (\frac{n_1}{n_2})^{\frac{n_1}{2}} x^{\frac{n_1}{2}-1} (1+\frac{n_1}{n_2}x)^{-\frac{n_1+n_2}{2}}, \quad x > 0$$

<hr>

```python
import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
fig, ax1 = plt.subplots(figsize = (7, 5))
xlim = [0, 4]
x = np.linspace(xlim[0], xlim[1], 1000)

# plot 1  n1、n2 互換差別不大，其中一個值固定，另一個值越大，越集中，且右偏
n1 = 60
n2 = np.arange(10, 81, 2)
colors = plt.colormaps['ocean'](np.linspace(0, 0.8, len(n2))) # 定义颜色列表
for pr, color in zip(n2, colors):
    y = f.pdf(x.reshape(-1, 1), dfn = pr, dfd = n1)
    ax1.plot(x, y, lw = 1.8, color = color, alpha = 0.8)
    ax1.tick_params(axis = 'both', colors = 'black')

fig.text(0.5, -0.01, '$x$', ha = 'center', fontsize = 12, color = 'black')
fig.text(-0.01, 0.5, '$Density$', va = 'center', rotation = 'vertical', fontsize = 12, color = 'black')
fig.text(0.6, 0.7, '$n_1 = 60,\ 10 \leq n_2 \leq 80$', fontsize = 12, color = 'black')
plt.suptitle(r'$F$ Distributions', fontsize = 15, fontweight = 'bold')
plt.tight_layout()
plt.show()
```
