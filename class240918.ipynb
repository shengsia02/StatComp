```python
# 老師的 code
# generate N samples from a normal distribution with mean 0 and variance 1
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

N = 10000
mu = 0
sigma = 1
x = norm.rvs(loc = mu, scale = sigma, size = N) # generate N samples from a normal distribution with mean mu and variance sigma
plt.figure(figsize=(6, 3))
plt.hist(x, edgecolor = 'orange', color = 'skyblue', bins = 10, density = True)
plt.xlabel('x')
plt.show()

# plot a pdf of the normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-5, 5, 100)
plt.figure(figsize=(6, 3))
plt.plot(x, norm.pdf(x, 0, 1))
plt.xlabel('x')
plt.ylabel('pdf')
plt.show()
```

```python
# generate N samples from a normal distribution with mean 0 and variance 1 and plot a pdf of the normal distribution on top of the histogram
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

N = 10000
mu = 0
sigma = 1
x = norm.rvs(loc = mu, scale = sigma, size = N) # generate N samples from a normal distribution with mean mu and variance sigma
plt.figure(figsize=(6, 3))
plt.hist(x, edgecolor = 'orange', color = 'skyblue', bins = 10, density = True)
plt.plot(x, norm.pdf(x, mu, sigma), 'r.', markersize = 1)
plt.xlabel('x')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
N = 100000  # Number of samples
mean = 0
sigma = 1

# Generate N samples from a normal distribution
samples = np.random.normal(mean, sigma, N)

# Create the histogram
plt.figure(figsize=(7, 5))
plt.hist(samples, bins = 30, density = True, alpha = 0.6, color = 'blue', edgecolor = 'r', label = 'Histrogram')

# Compute PDF values
x_values = np.linspace(-4, 4, 1000)  # Adjust the range as needed
pdf_values = norm.pdf(x_values, mean, sigma)

# Plot the PDF
plt.plot(x_values, pdf_values, 'g-', label = 'PDF', linewidth = 2, alpha = 0.6)

# Add labels and title
plt.xlabel('$\mu$')
plt.ylabel('Density')
plt.title('Histogram with Normal Distribution PDF')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
# make a 2x2 plots
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# generate N samples from a normal distribution with mean 0 and variance 1
N = 10000
mu = 0
sigma = 1
x = norm.rvs(loc = mu, scale = sigma, size = N) 
axs[0, 0].hist(x, edgecolor = 'orange', color = 'skyblue', bins = 10, density = True) # [0, 0] is topleft
axs[0, 0].set_xlabel('x')

# plot a pdf of the normal distribution
x = np.linspace(-5, 5, 100)
axs[0, 0].plot(x, norm.pdf(x, 0, 1))

# plot boxplot
x = np.random.normal(0, 1, 1000)
axs[0, 1].boxplot(x) # [0, 1] is topright
axs[0, 1].set_xlabel('x')

# plot qq plot
import scipy.stats as stats
stats.probplot(x, dist="norm", plot=axs[1, 0])
axs[1, 0].set_xlabel('Theoretical quantiles') # [1, 0] is bottomleft
axs[1, 0].set_ylabel('Ordered Values')

# plot a empirical cdf
x = np.random.normal(0, 1, 1000)
x = np.sort(x)
y = np.arange(1, len(x) + 1) / len(x)
axs[1, 1].plot(x, y) # [1, 1] is bottomright
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('CDF')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

dfs = np.arange(4, 32, 2)  # 自由度

# 定义颜色映射列表
colormaps = [
    'viridis', 'plasma', 'inferno', 'magma', 'cividis', 'Greys', 
    'Purples', 'Blues', 'Greens', 'Oranges', 'Reds', 'YlOrBr', 
    'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu', 'GnBu', 'PuBu', 
    'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn', 'PiYG', 'PRGn', 'BrBG', 
    'PuOr', 'RdGy', 'RdBu', 'RdYlBu', 'RdYlGn', 'Spectral', 
    'coolwarm', 'bwr', 'seismic', 'twilight', 'twilight_shifted', 
    'hsv', 'Pastel1', 'Pastel2', 'Paired', 'Accent', 'Dark2', 
    'Set1', 'Set2', 'Set3', 'tab10', 'tab20', 'tab20b', 'tab20c', 
    'flag', 'prism', 'ocean', 'gist_earth', 'terrain', 'gist_stern', 
    'gnuplot', 'gnuplot2', 'CMRmap', 'cubehelix', 'brg', 
    'gist_rainbow', 'rainbow', 'jet', 'nipy_spectral', 'gist_ncar'
]

fig, axes = plt.subplots(len(colormaps), 1, figsize=(10, 2 * len(colormaps)))

for ax, cmap in zip(axes, colormaps):
    colors = plt.cm.get_cmap(cmap)(np.linspace(0, 1, len(dfs)))
    for df, color in zip(dfs, colors):
        ax.plot([0, 1], [df, df], color=color, lw=6)
    ax.set_title(cmap)
    ax.set_yticks([])
    ax.set_xticks([])

plt.tight_layout()
plt.show()
```
