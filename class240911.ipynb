```python
# draw a pdf of a normal distribution
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math

# x = np.arange(-10, 10, 0.01)  # np.arange(start, stop, step), step 可設定間隔
x = np.linspace(-5, 5, 1000)  # np.linspace(start, stop, num), num 可設定生成的值的數量
y = stats.norm.pdf(x, loc = 0, scale = 1)  # loc = mean, scale = standard deviation

plt.figure(figsize = (6, 3))

plt.style.use('ggplot')
# plt.style.use('fivethirtyeight')
# plt.style.use('dark_background')

plt.plot(x, y, linewidth = 3, color = 'r', \
         label = 'Normal Distribution', alpha = 0.5, linestyle = '-')

plt.legend()
plt.grid(True)
plt.title('The Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
# generate N samples from a normal distribution with mean = 0 and standard deviation = 1 and plot

N = 1000000
# data = np.random.normal(0, 1, N)
data = stats.norm.rvs(loc = 0, scale = 1, size = N)

plt.figure(figsize = (7, 5))
plt.hist(data, bins = 200, color = 'b', alpha = 0.5, edgecolor = 'black')
plt.title('Histogram of the Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```
