```python
import numpy as np
import matplotlib.pyplot as plt

# 給定的四個數字
numbers = [2, 4, 9, 12]

# 抽樣次數
num_samples = 1000000

# 隨機抽取四個數字並計算其平均數
samples = np.random.choice(numbers, (num_samples, 4), replace=True)
averages = np.mean(samples, axis=1)

# 計算每個平均數的出現次數
unique_averages, counts = np.unique(averages, return_counts=True)

# 計算每個平均數的機率
pmf = counts / num_samples

# 檢查數據格式
print("Unique Averages:", unique_averages)
print("PMF:", pmf)

# 繪製 PMF
plt.figure(figsize=(10, 6))
plt.stem(unique_averages, pmf)
plt.xlabel('Average Value')
plt.ylabel('Probability')
plt.title('PMF of Random Variable Y')
plt.grid(True)
plt.show()
```

```python
# genenrate 100 samples from a normal distribution with mean = 0 and variance = 1

import numpy as np
from scipy.stats import norm
from scipy import stats
import matplotlib.pyplot as plt

N = 100
M = 10000
mu = 0
sigma = 1
P_Val = np.zeros([M, 1])
for i in range(M):
    x1 = norm.rvs(loc = mu, scale = sigma, size = N)
    x2 = norm.rvs(loc = mu, scale = sigma, size = N)
    # do a t-test and print the p-value
    results = stats.ttest_ind(x1, x2)
    P_Val[i] = results[1]

plt.hist(P_Val, bins = 100, density = True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, poisson, nbinom

# 設定 bin 參數
n = 50
p = 0.5
# 設定抽樣次數
N = 100
# 設定模擬次數
M = 10000
Xbar = np.zeros(M)

for i in range(M):
    X = binom.rvs(n, p, size = N)
    # 計算平均值
    Xbar[i] = np.mean(X)

# 計算每個平均數的出現次數
unique_X_bar, counts = np.unique(Xbar, return_counts = True)
# 計算每個平均數的機率
xbar_pmf = counts / M

# 計算理論常態分佈的 PDF，平均數為 n * p，變異數為 n * p * (1 - p) / N
N_norm = 100
mu = np.mean(Xbar)
sigma = np.std(Xbar)
x = np.linspace(np.min(unique_X_bar), np.max(unique_X_bar), N_norm)
y = norm.pdf(x, loc = mu, scale = sigma) / N_norm

# 繪畫二項分佈的 PMF
plt.figure(figsize=(10, 6))
# 繪製 PMF
plt.stem(unique_X_bar, xbar_pmf, basefmt=' ')
# 繪製理論常態分佈的 PDF
plt.plot(x, y, color = 'r')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('PMF of B({}, {})'.format(n, p))
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, poisson, nbinom

# 設定抽樣次數
N = 100
# 設定模擬次數
M = 10000

def CTL(Xbar):
    # 計算每個平均數的出現次數
    unique_X_bar, counts = np.unique(Xbar, return_counts = True)
    
    # 計算每個平均數的機率
    xbar_pmf = counts / M

    # 計算理論常態分佈的 PDF
    N_norm = 100
    mu = np.mean(Xbar)
    sigma = np.std(Xbar)
    x = np.linspace(np.min(unique_X_bar), np.max(unique_X_bar), N_norm)
    y = norm.pdf(x, loc = mu, scale = sigma) / N_norm
    return unique_X_bar, xbar_pmf, x, y


# 繪畫二項分佈的 PMF
fig, ax = plt.subplots(3, 1, figsize=(8, 11))

# Binomial Distribution
# 設定 bin 參數
n = 50
p = 0.5
Xbar1 = np.zeros(M)

for i in range(M):
    X1 = binom.rvs(n, p, size = N)
    # 計算平均值
    Xbar1[i] = np.mean(X1)
    
unique_X_bar, xbar_pmf, x, y = CTL(Xbar1)

# 繪製 bin PMF
ax[0].stem(unique_X_bar, xbar_pmf, basefmt=' ')
# 繪製理論常態分佈的 PDF
ax[0].plot(x, y, color = 'r')
ax[0].set_title('PMF of Bin({}, {})'.format(n, p))
ax[0].grid(True)


# Poisson Distribution
# 設定 Poi 參數
lamb = 5
Xbar2 = np.zeros(M)

for i in range(M):
    X2 = poisson.rvs(lamb, size = N)
    # 計算平均值
    Xbar2[i] = np.mean(X2)

unique_X_bar, xbar_pmf, x, y = CTL(Xbar2)

# 繪製 pois PMF
ax[1].stem(unique_X_bar, xbar_pmf, basefmt=' ')
# 繪製理論常態分佈的 PDF
ax[1].plot(x, y, color = 'r')
ax[1].set_title('PMF of Poi({})'.format(lamb))
ax[1].grid(True)


# NegativeBinomial Distribution
# 設定 Poi 參數
r = 5
p = 0.5
Xbar3 = np.zeros(M)

for i in range(M):
    X3 = nbinom.rvs(r, p, size = N)
    # 計算平均值
    Xbar3[i] = np.mean(X3)

unique_X_bar, xbar_pmf, x, y = CTL(Xbar3)

# 繪製 pois PMF
ax[2].stem(unique_X_bar, xbar_pmf, basefmt=' ')
# 繪製理論常態分佈的 PDF
ax[2].plot(x, y, color = 'r')
ax[2].set_title('PMF of NegaBin({}, {})'.format(r, p))
ax[2].grid(True)

plt.xlabel('Value')
# plt.ylabel('Probability')
plt.tight_layout()
plt.show()
```
